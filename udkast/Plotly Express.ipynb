```python
from datetime import datetime# Get today as string
today = datetime.now().strftime('%d-%m-%Y')# Get a data frame only for today
df_today = df[df.date == today]# Preview the data frame
df_today.head()
```

```python
import plotly.express as px

fig = px.scatter_geo(
    df_today, 
    locations='countryCode',
    color='continent',
    hover_name='country',
    size='cases',
    projection="natural earth",
    title=f'World COVID-19 Cases for {today}'
)
fig.show()
```
