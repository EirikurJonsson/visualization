```python
data = pd.read_csv("owid-covid-data.csv")
print(data.dtypes)
print(data.columns)
```

```python
data["date"] = pd.to_datetime(data["date"])
data = data.sort_values(by = "date")
data["total_cases_diff"] = data["total_cases"].diff()
```

```python
#df = data.transpose()
#df.columns = df.iloc[1]
#df = df.reset_index()
#df = df.drop([0,1,2,3])
#df = df.reset_index(drop = True)
data = data.fillna(0)
data["continent"] = data["continent"].astype('category')
#data["weekly_hosp_admissions"] = data["weekly_hosp_admissions"].astype('int')
```

```python
df = data.groupby(['date','location']).sum().fillna(0)
df.head()
```
