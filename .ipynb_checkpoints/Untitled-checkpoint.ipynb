```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv("https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv?raw=true")
```

```python
df = data.transpose()
df.columns = df.iloc[1]
df = df.reset_index()
df = df.drop([0,1,2,3])
df = df.reset_index(drop = True)
```

```python
df["date"] = pd.to_datetime(df["index"])
```

```python
fig = px.bar(df, x = "date", y = "Iceland")
fig.show()
```
